#1. Funciones
Hasta ahora hemos escrito código "lineal". Cuando el mismo patrón aparece varias veces, debemos usar funciones

###¿Para qué sirven?
* Evitan copiar/pegar
* Hacen el código más claro
* Facilitan cambios futuros
* Son clave para proyectos

sintaxis: -def nombre de la función-(elementos)

```python
def monto_con_igv(monto, igv):
    """
    esta funcion calcula el monto con igv
    input = monto(precio sin igv)
            IGV (0.18)
    output = precio con igv
    """
    monto_con_igv = monto * (1 + igv)
    return monto_con_igv #no todas las funciones retornan algo*
```

```python
monto_con_igv(100, 0.18)
```

```python
#Scope
#Las variables de la función no existen fuera de ella
x = 30
def f():
  x = 27
  return x

#¿por qué sucede esto?
'''
al definir x=30 fuera de la función, x se convierte en una variable global
al definir x=27 dentro de la función, x se convierte en una variable local
que solo existe dentro de esa función
entonces, cuando llamo a print(x), pythin busca x en el ámbito global, o sea =30
'''
```

#4. Gráficos con plt y seaborn
Los gráficos sirven para:
* detectar patrones
* detectar errores
* comunicar resultados

| Librería     | ¿Para qué se usa?                         | Cuándo usarla                         |
|-------------|-------------------------------------------|---------------------------------------|
| matplotlib  | Gráficos básicos y control total           | Líneas, histogramas, gráficos simples |
| seaborn     | Gráficos estadísticos y categóricos        | Barras, boxplots, comparaciones        |


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
summary_year = df.groupby("restaurant_year_opened")["restaurant_yelp_av_rating"].mean()
print(summary_year)
```

```python
#queremos graficar el mean por año
plt.figure()
plt.plot(summary_year.index, summary_year.values)
plt.show()
```

```python
plt.figure
plt.plot(summary_year.index, summary_year.values)
plt.title("Rating Promedio por Año")
plt.xlabel("Año")
plt.ylabel("Rating Promedio")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="restaurant_year_opened", y="restaurant_yelp_av_rating", data=df)
plt.title("Boxplot de Promedio de rating por año")
plt.xlabel("Año")
plt.ylabel("Rating")
plt.show()
```

```python
# resumen de estadísticos básicos por year
def resumen_stats_rating(df):
  summary_year = df.groupby("restaurant_year_opened")["restaurant_yelp_av_rating"].agg(["count", "mean", "std"]).reset_index()
  return summary_year
```

```python
resumen1 = resumen_stats_rating(df)
```
